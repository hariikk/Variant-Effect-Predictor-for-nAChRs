# important points
- state-of-the-art machine learning architectures such as Protein Language Models (PLMs) and Graph Neural Networks (GNNs)
- The human nAChR family is encoded by a multigene superfamily consisting of 16 distinct subunits: the $\alpha$-subunits ($\alpha$1–$\alpha$7, $\alpha$9, $\alpha$10), the $\beta$-subunits ($\beta$1–$\beta$4), and the muscle-specific $\gamma$, $\delta$, and $\varepsilon$ subunits. $\alpha$8 is in avian species and not humans.
- The assembly of nAChRs is governed by strict "compatibility rules," resulting in a finite set of functional combinations.

# Roadmap
# Phase I - Data Colleciton

- NAChRDB (The Primary Anchor): This is the single most valuable resource for this specific research endeavor. NAChRDB contains approximately 2,000 curated residue-level functional annotations derived from 117 studies. It aggregates data on mutagenesis, ligand binding, and electrophysiology. Crucially, it offers a Python-based query interface and allows data export in JSON format, making it machine-readable. The database covers 41 methods, including voltage-clamp and radioligand binding, providing a solid "seed" dataset of known critical residues. Actionable Step: Download the full JSON dump from NAChRDB to establish the baseline schema for the new dataset.

- FireProtDB (Stability Data): While focused on thermodynamic stability rather than channel gating, FireProtDB contains manually curated $\Delta\Delta G$ (Gibbs free energy change) values for single-point mutants. Stability is a major determinant of LoF (misfolded proteins are degraded), making this a vital orthogonal feature set. FireProtDB integrates data from ProTherm and ProtaBank, offering a clean, machine-learning-friendly dataset.

-ClinVar & gnomAD (The Clinical & Population Layer): ClinVar provides the clinical interpretation of variants, while gnomAD provides population allele frequencies. Variants with high frequency in gnomAD can serve as "Benign" functional controls. However, caution is advised: ClinVar labels often reflect clinical syndromes (e.g., Epilepsy) and do not explicitly state the biophysical mechanism (GoF vs. LoF). These labels should be treated as "distal" labels, distinct from the "proximal" labels derived from patch-clamp data
